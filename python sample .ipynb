You are working in a bank, and you have been given two lists of the employees who worked in 2021. Employees' names in list 1 are Ramesh, Suresh, Mahesh, Al, Jacob, and Santha List 2 contains the names of Al Mukesh Mahesh, Jacob, Sai, and Sanita. Please write a program that helps to identify people who are common in both lists. Please do not use any in-buit function

```python
# Define the two lists of employees
list1 = ['Ramesh', 'Suresh', 'Mahesh', 'Al', 'Jacob', 'Santha']
list2 = ['Al', 'Mukesh', 'Mahesh', 'Jacob', 'Sai', 'Sanita']

# Create an empty list to store common names
common_names = []

# Iterate through each name in list1
for name1 in list1:
    # Iterate through each name in list2
    for name2 in list2:
        # If the names match, add it to the common names list
        if name1 == name2:
            common_names.append(name1)

# Print the common names
if common_names:
    print("Common names in both lists:", common_names)
else:
    print("There are no common names in both lists.")
```

While entering dala, someone eritered a few names as a common string "Ramesh Suresh Mohit". Please write a program which separates all the names and converts them into a list. Once converted anto a list, please write a program that adds their age

Ramesh t 25 Suresh 22 Mohit 25

```python
# Given string containing names and ages
input_string = "Ramesh 25 Suresh 22 Mohit 25"

# Split the string into individual words
words = input_string.split()

# Initialize empty lists to store names and ages separately
names = []
ages = []

# Iterate over words in the list
for i in range(0, len(words), 2):  # Iterate over every second element (names)
    names.append(words[i])          # Append names to the names list

for i in range(1, len(words), 2):  # Iterate over every second element starting from index 1 (ages)
    ages.append(int(words[i]))      # Convert ages to integers and append to the ages list

# Print the separated names and ages
print("Names:", names)
print("Ages:", ages)

# Calculate the total age
total_age = sum(ages)

# Print the total age
print("Total age of all individuals:", total_age)
```

You are working in a medical store. A patient came to your medical store and asked to buy 2 strips of paracetamol, 3 strips of azithromycin, and 5 strips of Vitamin C. One strip of paracetamol costs Rs 36 one stop of azithromycin costs Rs 49 and one strip of vitamin c costs Rs. 33. Patient gave you Rs 2000. Please tell us what is the total cost of each medicine, the total cost of al mention and how much money you refunded to the patient

```python
# Define the prices of each medicine
price_paracetamol = 36
price_azithromycin = 49
price_vitamin_c = 33

# Define the quantities of each medicine
quantity_paracetamol = 2
quantity_azithromycin = 3
quantity_vitamin_c = 5

# Calculate the total cost of each medicine
total_cost_paracetamol = price_paracetamol * quantity_paracetamol
total_cost_azithromycin = price_azithromycin * quantity_azithromycin
total_cost_vitamin_c = price_vitamin_c * quantity_vitamin_c

# Calculate the total cost of all medicines
total_cost_all_medicines = (
    total_cost_paracetamol + total_cost_azithromycin + total_cost_vitamin_c
)

# Define the amount given by the patient
amount_given = 2000

# Calculate the total cost
total_cost = total_cost_all_medicines

# Calculate the amount to refund to the patient
refund_amount = amount_given - total_cost

# Print the total cost of each medicine
print("Total cost of Paracetamol:", total_cost_paracetamol)
print("Total cost of Azithromycin:", total_cost_azithromycin)
print("Total cost of Vitamin C:", total_cost_vitamin_c)

# Print the total cost of all medicines
print("Total cost of all medicines:", total_cost_all_medicines)

# Print the amount refunded to the patient
print("Refund amount to the patient:", refund_amount)
```

Accept a sentence as input and find the number of vowels in d. Assume that the sentence has no punctuation marks. For example. I am learning python contains vowels. This function should be applicable for all other different sentences


```python
def count_vowels(sentence):
    # Define a set of vowels
    vowels = {'a', 'e', 'i', 'o', 'u'}

    # Initialize a counter for vowels
    vowel_count = 0

    # Convert the sentence to lowercase to handle both uppercase and lowercase vowels
    sentence = sentence.lower()

    # Iterate through each character in the sentence
    for char in sentence:
        # If the character is a vowel, increment the vowel count
        if char in vowels:
            vowel_count += 1

    return vowel_count

# Test the function with a sample sentence
sentence = "I am learning python"
print("Number of vowels in the sentence:", count_vowels(sentence))
```

You have been appointed by the election commission to create a website. Your first task is to work on a program which tells candidates if they are eligible for voting or not. If they are give your output should be Congrats! You are eligible, otherwise, it should tell that you have to return after X number of years. The eligibility criteria for voting is 18 years


```python
def check_voting_eligibility(age):
    if age >= 18:
        print("Congrats! You are eligible.")
    else:
        years_to_wait = 18 - age
        print(f"Sorry, you have to return after {years_to_wait} years.")

# Test the function with different ages
candidate_age = int(input("Enter your age: "))
check_voting_eligibility(candidate_age)
```
